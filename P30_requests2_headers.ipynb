```python
import requests, re

class myCrawler():
    def __init__(self):
        pass
    
    def download(self, url):
        return (requests.get(url)).text # caution: .text is necessary
    
    def extruct(self, text_src, pattern):
        return re.findall(pattern, text_src)
    
    def save(self, file_addr, text_tar):
        with open(file_addr, 'w') as f: # 'a' or 'w'
            f.write(text_tar)
    
    def crawl(self, url, pattern, file_addr):
        content_src = self.download(url)
        content_list = self.extruct(content_src, pattern)
        content_tar = url + '\n\n'

        for item in content_list:
            for i in range(len(item)):
                content_tar += item[i]
                content_tar += ' ' if i < (len(item) - 1) else ''
            content_tar += '\n'
        
        self.save(file_addr, content_tar)
```

```python
crawler_zol = myCrawler()
crawler_zol.crawl('https://wap.zol.com.cn/top/cell_phone/hot.html', 
                 '<p class="pro-info-name f28">(.*?)<\/p>[\S\s]*?<span class="pro-info-price f24">(.*?)<\/span>', 
                 'C:\\Users\\mo\\jupyter_notebook_file\\P30_phones')
```
