```python
# 33-126 --> 33-126
def convert_char(single_char: str, operation: str) -> str:
    """转换单个字符
    输入：
    single_char：待转换字符
    operation： 'encrypt'解密 'decrypt'加密
    常数：
    OFFSET：ASCII码偏移量，只能正数，取值范围为[0, 93]
    返回：转换后字符
    """
    OFFSET = 3
    if ord(single_char) >= 33 and ord(single_char) <= 126: # in range(33, 127) 不易读
        if operation == 'encrypt':
            result = chr((ord(single_char) + OFFSET - 33) % (126 - 33 + 1) + 33)
        elif operation == 'decrypt':
            result = chr((ord(single_char) - OFFSET - 33) % (126 - 33 + 1) + 33)
    else:
        result = single_char
    return result

def encrypt_it(src_str: str) -> str:
    """字符串替换式加密
    使用了convert_char函数
    输入：
    src_char：待加密字符串
    返回：加密后字符串
    """
    tar_str = ''
    for single_char in src_str:
        tar_str += convert_char(single_char, 'encrypt')
    return tar_str

def decrypt_it(src_str: str) -> str:
    """字符串替换式解密
    使用了convert_char函数
    输入：
    src_char：待解密字符串
    返回：解密后字符串
    """
    tar_str = ''
    for single_char in src_str:
        tar_str += convert_char(single_char, 'decrypt')
    return tar_str

#验证加密解密结果
assert(decrypt_it(encrypt_it('abcdefghijklmnopqrstuvwxyzA!')) == 'abcdefghijklmnopqrstuvwxyzA!')
```

```python
# 生成一个随机码表，用dumps, loads函数
# 交付客户时，生成随机码表的函数不用给对方，给对方 key.dat 文件，对方进行loads操作

import random
import pickle

alphabet_src = [chr(i) for i in range(33, 127)] # append
alphabet_tar = alphabet_src.copy()
random.shuffle(alphabet_tar)

# f = open('P15_key.dat', mode='wb')
with open('P15_key.dat', mode='wb') as f: # 这样可以保证f的作用域仅在该语句内，且出该块时候会自动调用f.close()。wb表示二进制写，因为下面用到的dumps是将对象序列化并返回一个byte对象
    f.write(pickle.dumps([alphabet_src, alphabet_tar]))

#对方反序列化文件后可以得到码表
with open('P15_key.dat', mode='rb') as f: # rb表示二进制读
    print(pickle.loads(f.read()))
```
