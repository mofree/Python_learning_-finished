*P30中爬虫框架改进*  
*1. 增加self.headers变量*  

```python
import requests, re

class myCrawler():
    def __init__(self):
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36 Edg/83.0.478.56',
        } # 需要使用自定义headers，推荐进行update操作
    
    def download(self, url):
        return (requests.get(url, headers=self.headers)).text # caution: .text is necessary
    
    def extruct(self, text_src, pattern):
        return re.findall(pattern, text_src)
    
    def save(self, file_addr, text_tar):
        with open(file_addr, 'w') as f: # 'a' or 'w'
            f.write(text_tar)
    
    def crawl(self, url, pattern, file_addr):
        content_src = self.download(url)
        content_list = self.extruct(content_src, pattern)
        content_tar = url + '\n\n'

        for item in content_list:
            for i in range(len(item)):
                content_tar += item[i]
                content_tar += ' ' if i < (len(item) - 1) else ''
            content_tar += '\n'
        
        self.save(file_addr, content_tar)
```

```python
# 改进后可以读出信息了
crawler_douban = myCrawler()
crawler_douban.download('https://book.douban.com/tag/%E7%A7%91%E6%99%AE')
```
